```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('F:\Pandas dataset\diabetes.csv')

# Initial Inspection
print(df.head())
print(df.info())
print(df.describe())
```

```python
                                                      # Univariate Analysis
                                                      # Histogram for a continuous variable
sns.histplot(df['BMI'], kde=True)
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.show()
```

```python
sns.histplot(df['Glucose'], kde=True)
plt.show()
```

```python
sns.histplot(df['Pregnancies'], kde=True)
plt.show()
```

```python
                                                         # Boxplot for outliers
sns.boxplot(df['Pregnancies'])
plt.show()

sns.boxplot(df['SkinThickness'])
plt.show()

sns.boxplot(df['Glucose'])
plt.show()
```

```python
                                               # Bivariate Analysis
                                               # Scatter plot for two continuous variables
sns.scatterplot(x='Pregnancies', y='Age', data=df)
plt.show()
```

```python
sns.scatterplot(x='Glucose', y='Age', data=df)
plt.show()
```

```python
sns.scatterplot(x='Age', y='BloodPressure', data=df)
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
